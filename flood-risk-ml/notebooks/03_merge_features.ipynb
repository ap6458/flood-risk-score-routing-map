```python
import pandas as pd
rain_df=pd.read_csv("../data/processed/ward_rainfall_features.csv")
water_df=pd.read_csv("../data/processed/waterlogging_points.csv")
print(rain_df.head())
print(water_df.head())
```

```python
water_agg = water_df.groupby("ward").agg(
    waterlogging_count=("Severity", "count"),
    avg_severity=("Severity", "mean"),
    max_severity=("Severity", "max")
).reset_index()

water_agg.head(10)
```

```python
merged_df = rain_df.merge(
    water_agg,
    on="ward",
    how="left"
)
```

```python
merged_df["waterlogging_count"] = merged_df["waterlogging_count"].fillna(0)
merged_df["avg_severity"] = merged_df["avg_severity"].fillna(0)
merged_df["max_severity"] = merged_df["max_severity"].fillna(0)
```

```python
merged_df.info()
merged_df.head(10)
```

```python
merged_df.to_csv(
    "../data/processed/ward_flood_features1.csv",
    index=False
)
```
